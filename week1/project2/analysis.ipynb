```python
print("\n7. Summary Statistics")
print(df.describe())
```

```python
print("\n8. Statistical Analysis")

print("Total Students :", df["Student_ID"].count())
print("Average Age :", df["Age"].mean())
print("Average CGPA :", df["CGPA"].mean())

print("Median Age :", df["Age"].median())
print("Median CGPA :", df["CGPA"].median())

print("Minimum Age :", df["Age"].min())
print("Maximum Age :", df["Age"].max())

print("Minimum CGPA :", df["CGPA"].min())
print("Maximum CGPA :", df["CGPA"].max())
```

```python
print("\n9. Students with CGPA >= 3.7")

high_cgpa = df[df["CGPA"] >= 3.7]

print(high_cgpa)
```

```python
print("\n10. Students from Lahore having CGPA >= 3.7")

lahore_students = df[
    (df["City"] == "Lahore") &
    (df["CGPA"] >= 3.7)
]

print(lahore_students)
```

```python
print("\n11. Selected Columns")

selected_columns = df[["Name", "Department", "CGPA"]]

print(selected_columns)
```
