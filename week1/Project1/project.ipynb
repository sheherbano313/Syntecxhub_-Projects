```python
import numpy as np
import time

print("=" * 50)
print("NUMPY DATA EXPLORER")
print("=" * 50)
```

```python
print("\n1. ARRAY CREATION")

array1 = np.array([10, 20, 30, 40, 50])
array2 = np.arange(1, 13)
zeros = np.zeros((2, 3))
ones = np.ones((3, 3))
random_array = np.random.randint(1, 100, (3, 4))
```

```python
print("Array1:")
print(array1)
```

```python

print("\nArray2:")
print(array2)

```

```python
print("\nZeros:")
print(zeros)

```

```python
print("\nOnes:")
print(ones)
```

```python

print("\nRandom Array:")
print(random_array)
```

```python
# Indexing & Slicing
# -----------------------------
print("\n2. INDEXING & SLICING")

print("First Element:", array1[0])
print("Last Element:", array1[-1])
print("Slice:", array1[1:4])
```

```python
# Reshape
# -----------------------------
print("\n3. RESHAPING")

reshaped = array2.reshape(3, 4)

print(reshaped)
```

```python
print("\n4. MATHEMATICAL OPERATIONS")

print("Addition")
print(array1 + 5)

print("Multiplication")
print(array1 * 2)

print("Square")
print(np.square(array1))
```

```python
# Statistical Operations
# -----------------------------
print("\n5. STATISTICS")

print("Sum:", np.sum(array1))
print("Mean:", np.mean(array1))
print("Median:", np.median(array1))
print("Maximum:", np.max(array1))
print("Minimum:", np.min(array1))
print("Standard Deviation:", np.std(array1))
```

```python
print("\n6. AXIS OPERATIONS")

print("Row Sum:")
print(np.sum(reshaped, axis=1))

print("Column Sum:")
print(np.sum(reshaped, axis=0))
```

```python
print("\n7. BROADCASTING")

matrix = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

vector = np.array([10, 20, 30])

result = matrix + vector

print(result)
```

```python
# Save & Load
# -----------------------------
print("\n8. SAVE & LOAD")

np.save("saved_array.npy", reshaped)

loaded = np.load("saved_array.npy")

print("Loaded Array:")
print(loaded)
```

```python
print("\n9. PERFORMANCE TEST")

python_list = list(range(1000000))
numpy_array = np.arange(1000000)

start = time.time()
python_result = [x * 2 for x in python_list]
python_time = time.time() - start

start = time.time()
numpy_result = numpy_array * 2
numpy_time = time.time() - start

print(f"Python List Time : {python_time:.6f} sec")
print(f"NumPy Array Time : {numpy_time:.6f} sec")

if numpy_time < python_time:
    print("NumPy is Faster.")
else:
    print("Python List is Faster.")

print("\nProject Completed Successfully.")
```
