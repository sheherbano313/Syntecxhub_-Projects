```python
import pandas as pd

print("=" * 60)
print("        DATA CLEANING UTILITY")
print("=" * 60)
```

```python
df = pd.read_csv("raw_data.csv")

log = []

print("\nOriginal Dataset")
print(df)
```

```python
print("\nMissing Values")
print(df.isnull().sum())

log.append("Missing Values:")
log.append(str(df.isnull().sum()))
log.append("")
```

```python
df["Age"] = df["Age"].fillna(df["Age"].mean())
df["CGPA"] = df["CGPA"].fillna(df["CGPA"].mean())
```

```python
df["Join_Date"] = pd.to_datetime(
    df["Join_Date"],
    errors="coerce",
    format="mixed"
)

# Fill missing date with today's date
df["Join_Date"] = df["Join_Date"].fillna(pd.Timestamp.today())
```

```python
df["Age"] = df["Age"].astype(int)
df["CGPA"] = df["CGPA"].round(2)

print("\nData Types")
print(df.dtypes)

log.append("Data Types Fixed")
```

```python
duplicates = df.duplicated().sum()

df = df.drop_duplicates()

print(f"\nDuplicates Removed : {duplicates}")

log.append(f"Duplicates Removed : {duplicates}")
```

```python
df.columns = (
    df.columns
      .str.strip()
      .str.lower()
      .str.replace(" ", "_")
)

print("\nUpdated Column Names")
print(df.columns)

log.append("Column Names Standardized")
```

```python
df.to_csv("cleaned_data.csv", index=False)

log.append("Cleaned dataset saved as cleaned_data.csv")
```

```python
with open("cleaning_log.txt", "w") as file:
    for item in log:
        file.write(item + "\n")

print("\nCleaned Dataset")
print(df)

print("\nFiles Generated Successfully")
print("1. cleaned_data.csv")
print("2. cleaning_log.txt")

print("\nProject Completed Successfully.")
```
