```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("sales_data.csv")

df.head()
```

```python
df.info()
```

```python
plt.figure(figsize=(8,5))

plt.hist(df["Sales"], bins=15, edgecolor="black")

plt.title("Distribution of Sales")
plt.xlabel("Sales")
plt.ylabel("Frequency")

plt.tight_layout()

plt.savefig("sales_histogram.png")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.kdeplot(df["Sales"], fill=True)

plt.title("Sales Density (KDE)")
plt.xlabel("Sales")

plt.tight_layout()

plt.savefig("sales_kde.png")

plt.show()
```

```python
plt.figure(figsize=(7,5))

sns.boxplot(y=df["Sales"])

plt.title("Boxplot of Sales")

plt.tight_layout()

plt.savefig("sales_boxplot.png")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(x="Region", y="Sales", data=df)

plt.title("Sales Distribution by Region")

plt.tight_layout()

plt.savefig("region_comparison.png")

plt.show()
```

```python
print("Skewness:", df["Sales"].skew())
```

```python
Q1 = df["Sales"].quantile(0.25)
Q3 = df["Sales"].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = df[(df["Sales"] < lower) | (df["Sales"] > upper)]

print("Number of Outliers:", len(outliers))

print(outliers.head())
```
